### Analysis
 Study follows approximately 250 mice over 45 days to test efficacy of 9 cancer medications, with a placebo given to some mice as control. Capomulin is one of the most effective treatments for arresting tumor volume growth and reducing tumor volume over time. Figure 5 (below) shows this to be a reliable result throughout the study. In fact, all four selected treatments tested for reliability came back with consistent results, with only Infubinol having any outliers. Efficacy of treatment is adversely affected by weight of mouse (Figures 7 & 8). Figure 8 shows that this is a very strong correlation. Recommend testing higher dosage for heavier mice.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_study=pd.merge(mouse_metadata,study_results,how="left", on=["Mouse ID"])

# Display the data table for preview
mouse_study.head(20)
```

```python
#Rename age column 
mouse_study=mouse_study.rename(columns={"Age_months":"Age in Months"})
mouse_study.dtypes
```

```python
# Checking the number of mice.
mice=len(mouse_study["Mouse ID"].unique())
mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupes=mouse_study.loc[mouse_study.duplicated(["Mouse ID", "Timepoint"])]
dupes
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
gmouse=mouse_study[mouse_study["Mouse ID"] == "g989"]
gmouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#clean_mice=mouse_study[mouse_study["Mouse ID"].isin(dupes)==False]
#.drop_duplicates function?
clean_mice=mouse_study.loc[mouse_study["Mouse ID"]!="g989",:]
clean_mice
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_mice["Mouse ID"].unique())
```

```python
stats = clean_mice.describe()
stats
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimens=clean_mice.groupby("Drug Regimen")
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean=regimens["Tumor Volume (mm3)"].mean()
median=regimens["Tumor Volume (mm3)"].median()
variance=regimens["Tumor Volume (mm3)"].var()
stdev=regimens["Tumor Volume (mm3)"].std()
sem=regimens["Tumor Volume (mm3)"].sem()
#regimens.pd.describe(include='all').loc[[mean,median,variance,std,sem]]
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
summary_table=pd.DataFrame({"Mean Tumor Volume":mean,
                            "Median Tumor Volume":median,
                            "Tumor Volume Variance":variance,
                            "Tumor Volume Std. Dev": stdev,
                            "Tumor Volume Std. Error":sem})

summary_table
```

```python

# Using the aggregation method, produce the same summary statistics in a single line.
stats=clean_mice.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
stats.sort_index
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
#first_point=clean_mice.groupby(["Mouse ID"])["Timepoint"].min()
last_point=clean_mice.groupby(["Mouse ID"])["Timepoint"].max()
last_point=last_point.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_point_vol=last_point.merge(clean_mice, on=["Mouse ID","Timepoint"],how="left")
last_point_vol.head()
#sorted_df=last_point_vol.sort_values(by=["Tumor Volume (mm3)"])
#sorted_df.tail(50)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments=["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
volume_data=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:

     # Locate the rows which contain mice on each drug and get the tumor volumes
    final_vol=last_point_vol.loc[last_point_vol["Drug Regimen"]==treatment,"Tumor Volume (mm3)"]
    
    # add subset 
    volume_data.append(final_vol)
    
    # Determine outliers using upper and lower bounds
    quartiles = final_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = final_vol.loc[(final_vol < lower_bound) | (final_vol > upper_bound)]
    print(f" For {treatment}, the median final tumor volume is {round(quartiles[.5])}, the IQR is {round(iqr)}. Anything above {round(upper_bound)} or below {round(lower_bound)} can be considered an outlier.")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(volume_data, labels = treatments)
plt.ylabel('Final Tumor Volume (mm3)')
plt.title("Fig. 5- Final Tumor Volume for Selected Treatments")
plt.show()

```

## Correlation and Regression

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
average_cap=capomulin.groupby(["Mouse ID"]).mean()
plt.scatter(average_cap["Weight (g)"],average_cap["Tumor Volume (mm3)"])
plt.xlabel("Weight of Mouse")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Fig 7.- Weight of Mouse and Tumor Volume")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr=round(st.pearsonr(average_cap['Weight (g)'],average_cap['Tumor Volume (mm3)'])[0],2)
print(f'The correlation between weight and tumor volume is ' + (str(corr)))
model = st.linregress(average_cap['Weight (g)'],average_cap['Tumor Volume (mm3)'])
y_values = average_cap['Weight (g)']*model[0]+model[1]
plt.scatter(average_cap['Weight (g)'],average_cap['Tumor Volume (mm3)'])
plt.plot(average_cap['Weight (g)'],y_values,color="red")
#plt.title=("Relationship between Weight and Tumor Volume")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
print("This indicates a strong correlation between tumor volume and mouse weight.")
```
